```python
import pandas as pd
import rainflow
import matplotlib.pyplot as plt
filedemo='RejestratorCisnien11_05_23_12_43_07   c.csv'
#coldemo=['timestamp','delta']
```

```python
df=pd.read_csv(filedemo)
df.drop(df.index[0:1000], inplace=True)
df
```

```python
print(type(df))
```

```python
product = df['G-bok'].values.tolist()
product
```

```python
rnf = rainflow.count_cycles(product)
rnf
```

```python
print(type(rnf))
```

```python

x_val = [x[0] for x in rnf]
y_val = [x[1] for x in rnf]

#print(x_val)
plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
plt.show()
```

```python
fig, ax = plt.subplots(1,1) # gets a handle for the AxesSubplot object
x1 = ax.plot(x_val, y_val,)
x2 = ax.plot(x_val, y_val, 'b-')
#ax.set_yscale('log')
ax.set_xscale('log')
plt.title('Histogram zarejestrowanego obciÄ…Å¼enia')
plt.xlabel('napreÅ¼enia [MPa]')
plt.ylabel('libcza cykli')
fig.show()
```
